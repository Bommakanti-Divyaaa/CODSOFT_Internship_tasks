```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/Titanic_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# dropping unecssary columns
df = df.drop(['PassengerId', 'Name','Cabin','Ticket'], axis = 1)
df.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
# conerting into numbers
df["Sex"] = df["Sex"].map({
    'male': 0,
    'female': 1
})
df.head()
```

```python
df = pd.get_dummies(df, columns = ["Embarked"], drop_first = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/titanic_cleaned.csv", index = False)
```

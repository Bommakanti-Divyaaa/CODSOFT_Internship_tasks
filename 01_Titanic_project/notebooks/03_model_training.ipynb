```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../data/processed/titanic_cleaned.csv")
```

```python
# hiding the answers from the model while training.
x = df.drop("Survived", axis = 1)
y = df["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42
)
```

```python
model = LogisticRegression(max_iter = 1000)
model.fit(x_train, y_train)
```

```python
# predict
y_prid = model.predict(x_test)
y_prid
```

```python
acc = accuracy_score(y_test,y_prid)
print("Model accuracy: ", acc)
```

```python
results = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_prid
})
results.head(25)
```

```python
# checking training and testing accuracy
train_accuracy = model.score(x_train, y_train)
test_accuracy = model.score(x_test, y_test)
print(f"Train accuracy: {train_accuracy * 100:.2f}%" )
print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```

Model Evaluation

```python
# confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_prid)
print(cm)
```

```python
# plotting confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (6,4))

sns.heatmap(
    cm,
    annot =True,
    fmt = 'd',
    cmap = 'Blues'
)

plt.xlabel("Predicted")
plt.xlabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```

```python
# classification report
print(classification_report(y_test, y_prid))
```

```python
# saving the model
import joblib
joblib.dump(model, "../models/titanic_model.pkl")
joblib.dump(x.columns.tolist(), "../models/feature_names.pkl")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../dataset/raw/IMDb_movies_india.csv", encoding='latin1')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Name'])
```

```python
# removing rows without rating
df = df.dropna(subset=['Rating'])
print(df.shape)
```

```python
df['Year'].head()
```

```python
# formating the way year is written
df['Year'] = df['Year'].str.extract(r'(\d{4})')
df['Year'] = pd.to_numeric(df['Year'], errors = 'coerce')
df['Year'].head()
```

```python
# Force the column to a string type first, then strip the parentheses
df['Year'] = df['Year'].astype(str).str.strip('()')

# Convert it safely back to a number
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Year'].head()
```

```python
# cleaning duration column
df['Duration'] = df['Duration'].astype(str).str.extract(r'(\d+)')
df['Duration'] = pd.to_numeric(df['Duration'], errors = 'coerce')
df['Duration'].head()
```

```python
# cleaning votes column
df['Votes'] = df['Votes'].astype(str).str.replace(
    ',',
    '',
    regex = False
)
df['Votes'] = pd.to_numeric(df['Votes'], errors = 'coerce')
df['Votes'].head()
```

```python
# filling missing values - numerical
df['Year'] = df['Year'].fillna(
    df['Year'].median(),
    inplace = True
)

df['Duration'] = df['Duration'].fillna(
    df['Duration'].median(),
    inplace = True
)

df['Votes'] = df['Votes'].fillna(
    df['Votes'].median(),
    inplace=True
)
df.head()

```

```python
# filling missing values - Categorical
# genre, director, actor 1,2,3,
df['Genre'] = df['Genre'].fillna(
    'Unknown',
    inplace=True
)
df['Director'] = df['Director'].fillna(
    'Unknown',
    inplace=True
)

df['Actor 1'] = df['Actor 1'].fillna(
    'Unknown',
    inplace=True
)

df['Actor 2'] = df['Actor 2'].fillna(
    'Unknown',
    inplace=True
)

df['Actor 3'] = df['Actor 3'].fillna(
    'Unknown',
    inplace=True
)
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
# encoding categorical columns to intergers
genre_encoder = LabelEncoder()

df['Genre'] = genre_encoder.fit_transform(
    df['Genre']
)
df['Genre']
```

```python
#director
director_encoder = LabelEncoder()

df['Director'] = director_encoder.fit_transform(
    df['Director']
)
df['Director']
```

```python
# actor 1
actor1_encoder = LabelEncoder()

df['Actor 1'] = director_encoder.fit_transform(
    df['Actor 1']
)
df['Actor 1']
```

```python
# actor 2
actor2_encoder = LabelEncoder()

df['Actor 2'] = director_encoder.fit_transform(
    df['Actor 2']
)
df['Actor 2']
```

```python
# actor 3
actor3_encoder = LabelEncoder()

df['Actor 3'] = director_encoder.fit_transform(
    df['Actor 3']
)
df['Actor 3']
```

```python
# data types
df.info()
```

```python
# saving
df.to_csv("../dataset/processed/cleaned_movies.csv", index = False)
print("Preprocessed dataset saved successful!!")
```
